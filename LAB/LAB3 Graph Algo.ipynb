```python
#Task3
def action(state, action):
    new_state = {
        'M': state['M'] - action['M'],
        'C': state['C'] - action['C'],
        'boat': action['boat']
    }
    if not checkvalid(new_state):
        return None
    return new_state

def get_next_states(state):
    next_states = []
    for action in actions:
        if state['boat'] == 'left':
            if action['boat'] == 'right':
                continue
            new_state = action(state, action)
        else:
            if action['boat'] == 'left':
                continue
            new_state = action(state, action)
        if new_state is not None:
            next_states.append(new_state)
    return next_states

def goal(state):
    return state == goal

def bfs(initial):
    queue = initial
    visited = set()
    while queue:
        path = queue.pop(0)
        node = path[-1]
        if goal(node):
            return path
        for state in get_next_states(node):
            if tuple(state.items()) not in visited:
                visited.add(tuple(state.items()))
                new_path = list(path)
                new_path.append(state)
                queue.append(new_path)
    return None

def checkvalid(state):
    if state['M'] < 0 or state['C'] < 0:
        return False
    if state['M'] > 3 or state['C'] > 3:
        return False
    if state['C'] > state['M'] and state['M'] > 0:
        return False
    if state['C'] < state['M'] and state['M'] < 3:
        return False
    return True

initial = {'M': 3, 'C': 3, 'boat': 'left'}
goal = {'M': 0, 'C': 0, 'boat': 'right'}

actions = [
    {'M': 2, 'C': 0, 'boat': 'right'},
    {'M': 1, 'C': 0, 'boat': 'right'},
    {'M': 1, 'C': 1, 'boat': 'right'},
    {'M': 0, 'C': 1, 'boat': 'right'},
    {'M': 0, 'C': 2, 'boat': 'right'},
    {'M': 1, 'C': 1, 'boat': 'left'},
    {'M': 2, 'C': 2, 'boat': 'left'},
    {'M': 2, 'C': 1, 'boat': 'left'},
    {'M': 1, 'C': 0, 'boat': 'left'}
]





solution = bfs(initial)
for state in solution:
    print(state)
```
